```python
# profile1_1

import matplotlib.pyplot as plt

dat1_1 = []

i=0;

with open('profile1.data') as inputfile:
    for line in inputfile:
        if(line[0]!='0'):
            dat1_1.append(line.strip().split(' '))
time=[]
page_fault=[]
pf_sum=0

time_0=int(dat1_1[0][0])
for i in  range(len(dat1_1)-1):
    for j in range(4):
        dat1_1[i][j]=int(dat1_1[i][j])
    time.append(dat1_1[i][0]-time_0)
    pf_sum=pf_sum+dat1_1[i][1]+dat1_1[i][2]
    page_fault.append(pf_sum)
        
plt.plot(time, page_fault)
plt.xlabel('Time /  jiffies') 
# naming the y axis 
plt.ylabel('Page fault') 
# giving a title to my graph 
plt.title('The page fault with time for case1.1')
```
